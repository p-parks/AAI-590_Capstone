```python
from transformers import pipeline, T5ForConditionalGeneration, T5Tokenizer, BartForConditionalGeneration, BartTokenizer
```

```python
# First Model
# Load T5 model and tokenizer
t5_model = T5ForConditionalGeneration.from_pretrained("t5-base")
t5_tokenizer = T5Tokenizer.from_pretrained("t5-base")

# Initialize the summarizer pipeline
summarizer_t5 = pipeline("summarization", model=t5_model, tokenizer=t5_tokenizer)
```

```python
# Second Model
# Load BART model and tokenizer
bart_model = BartForConditionalGeneration.from_pretrained("facebook/bart-large-cnn")
bart_tokenizer = BartTokenizer.from_pretrained("facebook/bart-large-cnn")

# Initialize the summarizer pipeline
summarizer_bart = pipeline("summarization", model=bart_model, tokenizer=bart_tokenizer)
```

```python
import pandas as pd

#Dataset
podcast_data = pd.read_csv('/content/podcastdata_dataset.txt', delimiter="\t")

# Display the first few rows of the data
print(podcast_data.head())
```

```python
# Function to summarize the transcript
def summarize_transcript(text, model='t5'):
    if model == 't5':
        summary = summarizer_t5(text, max_length=50, min_length=25, do_sample=False)
    else:
        summary = summarizer_bart(text, max_length=50, min_length=25, do_sample=False)
    return summary[0]['summary_text']

last_column_name = podcast_data.columns[-1]
podcast_data['summary'] = podcast_data[last_column_name].apply(summarize_transcript, model='bart')

# Display the summarized data
print(podcast_data[[last_column_name, 'summary']].head())
```
